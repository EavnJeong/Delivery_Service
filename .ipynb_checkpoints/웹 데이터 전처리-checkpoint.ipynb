```python
import pandas as pd
import os
```

```python
rest = pd.read_table('rest.txt')
price = pd.read_csv('restprice.csv')
```

```python
rest = rest.drop_duplicates('name')

rest.reset_index(inplace=True)

rest = rest.iloc[:, 1:]

price = price.sort_values('id')
price.reset_index(inplace=True)
price = price.iloc[:, 1:]
price

rest = rest.sort_values('id')
rest.reset_index(inplace=True)
rest = rest.iloc[:, 1:]
```

```python
rest['price'] = 0
rest['menu'] = 0

rest['price'] = rest['price'].astype(object)
rest['menu'] = rest['menu'].astype(object)
```

```python
import numpy as np
rest2 = np.array(rest['id'])
price2 = np.array(price['id'])

boo = []
for i in range(len(rest2)):
    boo.append(rest2[i] in price2)
boo2 = []
for i in range(len(price2)):
    boo2.append(price2[i] in rest2)

rest = rest[boo]
price = price[boo2]

rest.reset_index(inplace=True)
rest = rest.iloc[:, 1:]

price.reset_index(inplace=True)
price = price.iloc[:, 1:]
```

```python
prices = []
foods = []

j = 0

for i in range(158702):

    if price.iloc[i, 0] == price.iloc[i+1, 0]:
        foods.append(price.iloc[i, 1])
        prices.append(price.iloc[i, 2])
    else:
        rest.at[j, 'menu'] = foods
        rest.at[j, 'price'] = prices
        prices = []
        foods = []
        j = j + 1
```

```python
rest = rest[rest['price'] != 0]
```

```python
conv = pd.read_table('conv.txt')
convprice = pd.read_table('convprice.txt')
```

```python
conv = conv.drop_duplicates('name')
conv.reset_index(inplace=True)

conv = conv.iloc[:, 1:]

convprice = convprice.sort_values('id')
convprice.reset_index(inplace=True)
convprice = convprice.iloc[:, 1:]
convprice

conv = conv.sort_values('id')
conv.reset_index(inplace=True)
conv = conv.iloc[:, 1:]
```

```python
conv['price'] = 0
conv['menu'] = 0

conv['price'] = conv['price'].astype(object)
conv['menu'] = conv['menu'].astype(object)
```

```python
import numpy as np
rest2 = np.array(conv['id'])
price2 = np.array(convprice['id'])

boo = []
for i in range(len(rest2)):
    boo.append(rest2[i] in price2)
boo2 = []
for i in range(len(price2)):
    boo2.append(price2[i] in rest2)

conv = conv[boo]
convprice = convprice[boo2]

conv.reset_index(inplace=True)
conv = conv.iloc[:, 1:]

convprice.reset_index(inplace=True)
convprice = convprice.iloc[:, 1:]
```

```python
prices = []
foods = []

j = 0
for i in range(10579):
    if convprice.iloc[i, 1] == convprice.iloc[i+1, 1]:
        foods.append(convprice.iloc[i, 2])
        prices.append(convprice.iloc[i, 3])
    else:
        conv.at[j, 'menu'] = foods
        conv.at[j, 'price'] = prices
        prices = []
        foods = []
        j = j + 1
```

```python
conv = conv.dropna()
conv = conv[conv['price']!=0]
conv = conv.iloc[:, 1:]
conv
```
